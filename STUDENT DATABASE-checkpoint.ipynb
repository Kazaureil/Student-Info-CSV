```python
from tkinter import *
import csv
from tkinter import messagebox as ms
import uuid

# Setting this as global so it can be accessed and be manipulated all throughout the body
global data
global row
data = []
with open('Infos.csv', 'r') as rf:
    n = csv.reader(rf)
    for row in n:
        if row:
            data.append(row)


# Main Class which consist of all operations and codes for the GUI
class main:
    def __init__(self, master):
        self.master = master
        self.studentID = StringVar()
        self.fullname = StringVar()
        self.email = StringVar()
        self.number = StringVar()
        self.var = StringVar()
        self.optionList = ["Mobile", "Work", "Home"]
        self.mobile = StringVar()
        self.work = StringVar()
        self.home = StringVar()
        
        
        self.n_studentID = StringVar()
        self.n_fullname = StringVar()
        self.n_email = StringVar()
        self.n_number = StringVar()
        self.n_var = StringVar()
        self.n_mobile = StringVar()
        self.n_work = StringVar()
        self.n_home = StringVar()
        # Create Widgets
        self.widgets()

    # This Function is used to search and verify if a contact exists
    def search_contacts(self):
        mirror = bool
        studentID = self.n_studentID.get()
        for row in data:
            for field in row:
                if field == studentID:
                    mirror = True
                    print(row)
                    print(data)
                    #StudentID
                    Label(self.find, text='Student ID: ', font=('', 15), pady=5, padx=5).grid(row=4, column=0)
                    name = Label(self.find, text=str(row[0]), font=('', 15))
                    name.grid(row=4, column=1)
                    
                    # fullname
                    Label(self.find, text='Full Name: ', font=('', 15), pady=5, padx=5).grid(row=5, column=0)
                    name = Label(self.find, text=str(row[1]), font=('', 15))
                    name.grid(row=5, column=1)
                    Button(self.find, text='EDIT/CONFIRM', bd=3, font=('', 10), padx=5, command=self.UpdateName).grid(row=5,
                                                                                                              column=2)

                    # email
                    Label(self.find, text='E-MAIL: ', font=('', 15), pady=5, padx=5).grid(row=6, column=0)
                    email = Label(self.find, text=row[2], font=('', 15))
                    email.grid(row=6, column=1)
                    Button(self.find, text='EDIT/CONFIRM', bd=3, font=('', 10), padx=5, command=self.UpdateEmail).grid(row=6,
                                                                                                              column=2)

                    # mobile
                    Label(self.find, text='MOBILE NUMBER: ', font=('', 15), pady=5, padx=5).grid(row=7, column=0)
                    mobile = Label(self.find, text=row[3], font=('', 15))
                    mobile.grid(row=7, column=1)
                    Button(self.find, text='EDIT/CONFIRM', bd=3, font=('', 10), padx=5, command=self.UpdateMobile).grid(row=7,
                                                                                                                column=2)
                    # work
                    Label(self.find, text='WORK NUMBER: ', font=('', 15), pady=5, padx=5).grid(row=8, column=0)
                    work = Label(self.find, text=row[4], font=('', 15))
                    work.grid(row=8, column=1)
                    Button(self.find, text='EDIT/CONFIRM', bd=3, font=('', 10), padx=5, command=self.UpdateWork).grid(row=8,
                                                                                                              column=2)

                    # home
                    Label(self.find, text='HOME NUMBER: ', font=('', 15), pady=5, padx=5).grid(row=9, column=0)
                    home = Label(self.find, text=row[5], font=('', 15))
                    home.grid(row=9, column=1)
                    Button(self.find, text='EDIT/CONFIRM', bd=3, font=('', 10), padx=5, command=self.UpdateHome).grid(row=9,
                                                                                                              column=2)
                    Button(self.find, text='UPDATE', bd=3, font=('', 15), padx=50, command=self.update_writer).grid(
                        row=10, column=0)
                    Button(self.find, text='DELETE', bd=3, font=('', 15), padx=50, command=self.Delete).grid(row=10,
                                                                                                            column=1)
        if mirror != True:
            ms.showerror('Error', 'ID Not Found!')
            
            
    def UpdateStudID(self):
        studentID = self.n_studentID.get()
        for row in data:
            for field in row:
                if field == studentID:
                    Entry(self.find, textvariable=self.n_studentID, bd=5, font=('', 15)).grid(row=4, column=1)
                    x = int()
                    mirror = bool
                    student_ID = self.n_studentID.get()
                    for row in data:
                        for field in row:
                            if field == self.studentID.get():
                                mirror = True
                    if mirror == True:
                        ms.showerror('Oops!', 'An information with the same name Already exists')
                    else:
                        while x != 1:
                            for row in data:
                                for field in row:
                                    if field == student_ID:
                                        row[0] = student_ID
                        with open('Infos.csv', 'w') as wf:
                            write_data = csv.writer(wf)
                            for line in data:
                                write_data.writerow(line)
                                
    # This Function is used Edit the Name of the contact
    def UpdateName(self):
        studentID = self.n_studentID.get()
        for row in data:
            for field in row:
                if field == studentID:
                    Entry(self.find, textvariable=self.n_fullname, bd=5, font=('', 15)).grid(row=5, column=1)
                    x = int()
                    student_ID = self.n_studentID.get()
                    full_name = self.n_fullname.get()
                    while x != 1:
                        for row in data:
                            for field in row:
                                if field == student_ID:
                                    row[1] = full_name
                                    x = 1
                    with open('Infos.csv', 'w') as wf:
                        write_data = csv.writer(wf)
                        for line in data:
                            write_data.writerow(line)

    # This Function is used Edit the Name of the contact
    def UpdateEmail(self):
        studentID = self.n_studentID.get()
        for row in data:
            for field in row:
                if field == studentID:
                    Entry(self.find, textvariable=self.n_email, bd=5, font=('', 15)).grid(row=6, column=1)
                    x = int()
                    student_ID = self.n_studentID.get()
                    email_info = self.n_email.get()
                    while x != 1:
                        for row in data:
                            for field in row:
                                if field == student_ID:
                                    row[2] = email_info
                                    x = 1
                    with open('Infos.csv', 'w') as wf:
                        write_data = csv.writer(wf)
                        for line in data:
                            write_data.writerow(line)

    # This Function is used Edit the Name of the contact
    def UpdateMobile(self):
        studentID = self.n_studentID.get()
        for row in data:
            for field in row:
                if field == studentID:
                    Entry(self.find, textvariable=self.n_mobile, bd=5, font=('', 15)).grid(row=7, column=1)
                    x = int()
                    student_ID = self.n_studentID.get()
                    mobile = self.n_mobile.get()
                    while x != 1:
                        for row in data:
                            for field in row:
                                if field == student_ID:
                                    row[3] = mobile
                                    x = 1
                    with open('Infos.csv', 'w') as wf:
                        write_data = csv.writer(wf)
                        for line in data:
                            write_data.writerow(line)

    # This Function is used Edit the Name of the contact
    def UpdateWork(self):
        studentID = self.n_studentID.get()
        for row in data:
            for field in row:
                if field == studentID:
                    Entry(self.find, textvariable=self.n_work, bd=5, font=('', 15)).grid(row=8, column=1)

                    x = int()
                    student_ID = self.n_studentID.get()
                    work = self.n_work.get()
                    while x != 1:
                        for row in data:
                            for field in row:
                                if field == student_ID:
                                    row[4] = work
                                    x = 1
                    with open('Infos.csv', 'w') as wf:
                        write_data = csv.writer(wf)
                        for line in data:
                            write_data.writerow(line)

    # This Function is used Edit the Name of the contact
    def UpdateHome(self):
        studentID = self.n_studentID.get()
        for row in data:
            for field in row:
                if field == studentID:
                    Entry(self.find, textvariable=self.n_home, bd=5, font=('', 15)).grid(row=9, column=1)
                    x = int()
                    student_ID = self.n_studentID.get()
                    home = self.n_home.get()
                    while x != 1:
                        for row in data:
                            for field in row:
                                if field == student_ID:
                                    row[5] = home
                                    x = 1
                    with open('Infos.csv', 'w') as wf:
                        write_data = csv.writer(wf)
                        for line in data:
                            write_data.writerow(line)

    # This Function is used to write the manipulated database list to an external database
    def update_writer(self):
        ms.showinfo('Success!', 'Contact Updated')
        with open('Infos.csv', 'w') as wf:
            write_data = csv.writer(wf)
            for line in data:
                write_data.writerow(line)
     

        print(data)

    # This Function is used to Delete a certain contact
    def Delete(self):
        student_ID = self.n_studentID.get()
        for row in data:
            for field in row:
                if student_ID == field:
                    data.remove(row)
                    with open('Infos.csv', 'w') as wf:
                        write_data = csv.writer(wf)
                        for line in data:
                            write_data.writerow(line)
                    print(data)
        ms.showinfo('Success!', 'Information Deleted')
        with open('Infos.csv', 'w') as wf:
            write_data = csv.writer(wf)
            for line in data:
                write_data.writerow(line)
       

    # This Function is used to add Contacts and to append a certain data to the global variable
    def add_contacts(self):
        print(data)
        mobile = str()
        work = str()
        home = str()
        mirror = bool
        studentID = self.studentID.get()
        for row in data:
            for field in row:
                if field == self.studentID.get():
                    mirror = True
        if mirror == True:
            ms.showerror('Oops!', 'An info with the same Name/ID number Already exists')
        else:
            email = self.email.get()
            fullname = self.fullname.get()
            type = self.var.get()
            id = uuid.uuid4()
            print(type)
            if type == 'Mobile':
                mobile = self.number.get()
                print(mobile)
                work = 'None'
                home = 'None'
            elif type == 'Work':
                work = self.number.get()
                print(work)
                mobile = 'None'
                home = 'None'
            elif type == 'Home':
                home = self.number.get()
                print(home)
                mobile = 'None'
                work = 'None'
            with open('Infos.csv', 'a') as rf:
                fieldnames = ['STUDENT_ID','FULL_NAME', 'EMAIL', 'MOBILE', 'WORK', 'HOME','UNIQUE_ID']
                n = csv.DictWriter(rf, fieldnames=fieldnames)
                n.writerow(
                    {'STUDENT_ID': studentID, 'FULL_NAME': fullname, 'EMAIL': email,  'MOBILE': mobile, 'WORK': work, 'HOME': home, 'UNIQUE_ID': id })
        

            ms.showinfo('Success!', 'Contact Added')
       

    # This Functions are used to setup the Packing methods of the widgets for the GUI
    def main(self):
        self.n_fullname.set('')
        self.n_email.set('')
        self.n_studentID.set('')
        self.create.pack_forget()
        self.find.pack_forget()
        self.head['text'] = 'HOME'
        self.home.pack()

    def add(self):
        self.n_studentID.set('')
        self.n_fullname.set('')
        self.n_email.set('')
        self.create.pack_forget()
        self.home.pack_forget()
        self.find.pack_forget()
        self.head['text'] = 'ADD A NEW STUDENT INFO'
        self.create.pack()

    def search(self):
        self.n_studentID.set('')
        self.n_fullname.set('')
        self.n_email.set('')
        self.home.pack_forget()
        self.create.pack_forget()
        self.head['text'] = 'SEARCH FOR STUDENT INFO'
        self.find.pack()
        
    def show(self):
        self.n_studentID.set('')
        self.n_fullname.set('')
        self.n_email.set('')
        self.find.pack_forget()
        self.home.pack_forget()
        self.create.pack_forget()
        self.head['text'] = 'UPDATE STUDENT INFO'
        self.show.pack()

    # This is used to setup the interface
    def widgets(self):
        # entry part
        self.head = Label(self.master, text="HOME", font=('', 20))
        self.head.pack()
        Label(self.master, text="").pack()

        ##------- HOME --------##
        self.home = Frame(self.master, padx=10, pady=1, )
        # buttons part
        Button(self.home, text=' SEARCH FOR STUDENT INFO', bd=3, font=('', 12), padx=5, pady=5, command=self.search).pack()
        Label(self.home, text="").pack()
        Button(self.home, text='   ADD A NEW STUDENT INFO  ', bd=3, font=('', 12), padx=5, pady=5, command=self.add).pack()
        Label(self.home, text="").pack()
        self.home.pack()
        ##------- ADD CONTACTS --------##
        self.create = Frame(self.master, padx=10, pady=10)
        Label(self.create, text='Student ID: ', font=('', 15), pady=5, padx=5).grid(sticky=W, row=2, column=0)
        Entry(self.create, textvariable=self.studentID, bd=5, font=('', 15)).grid(row=2, column=1)
        
        Label(self.create, text='Full Name: ', font=('', 15), pady=5, padx=5).grid(sticky=W)
        Entry(self.create, textvariable=self.fullname, bd=5, font=('', 15)).grid(row=3, column=1)

        Label(self.create, text='E-mail: ', font=('', 15), pady=5, padx=5).grid(sticky=W)
        Entry(self.create, textvariable=self.email, bd=5, font=('', 15)).grid(row=4, column=1)

        Label(self.create, text='Number: ', font=('', 15), pady=5, padx=5).grid(sticky=W)
        Label(self.create, text='Type: ', font=('', 15), pady=5, padx=5).grid(sticky=W)
        
        OptionMenu(self.create, self.var, *self.optionList).grid(row=6, columnspan=2)
        Entry(self.create, textvariable=self.number, bd=5, font=('', 15)).grid(row=5, column=1)
        

        Button(self.create, text='Add Contact', bd=3, font=('', 15), padx=30, pady=5, command=self.add_contacts).grid()
        Button(self.create, text='Back', bd=3, font=('', 15), padx=70, pady=5, command=self.main).grid(
            row=7, column=1)
        ##------- SEARCH --------##
        self.find = Frame(self.master, padx=10, pady=10)
        Label(self.find, text='Student ID: ', font=('', 15), pady=5, padx=5).grid(sticky=W, row=2, column=0)
        Entry(self.find, textvariable=self.n_studentID, bd=5, font=('', 15)).grid(row=2, column=1)
        Button(self.find, text='Back', bd=3, font=('', 15), padx=50, pady=5, command=self.main).grid()
        Button(self.find, text='Search', bd=3, font=('', 15), padx=75, pady=5, command=self.search_contacts).grid(row = 3,column=1)
       

# Runs the GUI
root = Tk()
main(root)
root.title("Contacts")
root.mainloop()
```
